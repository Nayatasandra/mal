```python
# Isi missing values (jika ada)
data.fillna(data.mean(), inplace=True)
```

```python
# Normalisasi data (kecuali kolom 'Nama')
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data.drop(columns=['Nama']))
# simpan scaled_data
joblib.dump(scaled_data, 'scaled_data.pkl')
joblib.dump(scaler, 'scaler.pkl')
```

```python
# Tampilkan data yang sudah dinormalisasi
print("\nData setelah normalisasi (5 baris pertama):")
print(pd.DataFrame(scaled_data, columns=data.columns[1:]).head())
```

```python
from pyclustering.cluster.cure import cure
from pyclustering.cluster import cluster_visualizer_multidim
from pyclustering.samples.definitions import SIMPLE_SAMPLES

# Membuat model CURE
cure_instance = cure(scaled_data.tolist(), number_cluster=3, number_represent_points=5, compression=0.5)
cure_instance.process()

# Mendapatkan hasil klaster
clusters = cure_instance.get_clusters()
```

```python
# Visualisasi untuk data multidimensi
visualizer = cluster_visualizer_multidim()
visualizer.append_clusters(clusters, scaled_data.tolist())
# visualizer.show()
```

```python
# Simpan model
joblib.dump(cure_instance, 'cure_model.pkl')
print("Model CURE telah disimpan ke cure_model.pkl")
```

```python
from sklearn.metrics import silhouette_score

# Ambil data dan kluster yang dihasilkan
data = scaled_data  # Data yang di-cluster
clusters = cure_instance.get_clusters()

# Buat array yang mewakili setiap kluster untuk data
cluster_labels = np.zeros(len(data))

for cluster_idx, cluster in enumerate(clusters):
    for index in cluster:
        cluster_labels[index] = cluster_idx

# Hitung Silhouette Score
silhouette_avg = silhouette_score(data, cluster_labels)
print(f"Silhouette Score: {silhouette_avg}")
```

```python
from scipy.spatial.distance import pdist, squareform
from scipy.spatial.distance import cdist

# Menghitung Dunn Index
def dunn_index(data, labels):
    distances = squareform(pdist(data))
    unique_labels = np.unique(labels)
    
    intra_distances = []
    inter_distances = []
    
    for label in unique_labels:
        cluster_points = data[labels == label]
        other_points = data[labels != label]
        
        # Intra-cluster distances (jarak dalam kluster)
        if len(cluster_points) > 1:
            intra_distances.append(np.max(pdist(cluster_points)))
        
        # Inter-cluster distances (jarak antar-kluster)
        for other_label in unique_labels:
            if label != other_label:
                other_cluster_points = data[labels == other_label]
                inter_distances.append(np.min(cdist(cluster_points, other_cluster_points)))
    
    dunn = np.min(inter_distances) / np.max(intra_distances)
    return dunn

# Hitung Dunn Index
dunn = dunn_index(data, cluster_labels)
print(f"Dunn Index: {dunn}")
```

```python
from sklearn.metrics import davies_bouldin_score

db_score = davies_bouldin_score(data, cluster_labels)
print(f"Davies-Bouldin Index: {db_score}")
```

```python
import joblib
import pandas as pd
from scipy.spatial.distance import cdist

# Muat data baru
data_baru = pd.DataFrame({
    'Nama': ['Hasan'],
    'Umur': [75],
    'Jenis_Kelamin': ['Laki-laki'],
    'LiLA': [30],
    'Berat_Badan': [50],
    'Tinggi_Badan': [160],
    'Penurunan_Berat_Badan':[0],
    'Frekuensi_Makan': ['Dua kali'],
    'Variasi_Makanan': ['Makanan bervariasi (misalnya daging, sayuran, buah)'],
    'Asupan_Protein': ['Dua kali sehari atau lebih'],
    'Mobilitas': ['Bergerak dengan bebas tanpa bantuan'],
    'Aktivitas_Sehari_hari': ['Tidak ada kesulitan'],
    'Kesehatan_Mulut': ['Tidak ada masalah'],
    'Penyakit_Kronis': ['Tidak'],
    'Stres': ['Tidak'],
    'Masalah_Kognitif': ['Ada sedikit masalah daya ingat']
})

# Preprocessing data baru (ubah kategorikal dan normalisasi)
def preprocess_data(data):
    data['Jenis_Kelamin'] = data['Jenis_Kelamin'].map({'Perempuan': 1, 'Laki-laki': 0})
    data['Penyakit_Kronis'] = data['Penyakit_Kronis'].map({'Ya': 1, 'Tidak': 0})
    data['Stres'] = data['Stres'].map({'Ya': 1, 'Tidak': 0})
    data['Frekuensi_Makan'] = data['Frekuensi_Makan'].map({
        'Tiga kali atau lebih':3,
        'Dua kali': 2,
        'Satu kali atau kurang': 1
    })
    
    data['Variasi_Makanan'] = data['Variasi_Makanan'].map({
        'Makanan bervariasi (misalnya daging, sayuran, buah)': 3,
        'Makanan terbatas pada beberapa jenis saja': 2,
        'Makanan sangat terbatas atau monoton': 1
    })
    
    data['Asupan_Protein'] = data['Asupan_Protein'].map({
        'Dua kali sehari atau lebih': 3,
        'Sekali sehari': 2,
        'Kurang dari sekali sehari atau tidak pernah':1
    })
    
    data['Mobilitas']=data['Mobilitas'].map({
        'Bergerak dengan bebas tanpa bantuan':3,
        'Bergerak dengan bantuan':2,
        'Tidak dapat bergerak tanpa bantuan orang lain':1
    })
    
    data['Aktivitas_Sehari_hari']=data['Aktivitas_Sehari_hari'].map({
        'Tidak ada kesulitan':3,
        'Kesulitan ringan':2,
        'Kesulitan berat':1
    })
    
    data['Kesehatan_Mulut']=data['Kesehatan_Mulut'].map({
        'Tidak ada masalah':4,
        'Ada kesulitan ringan':3,
        'Kesulitan berat':2,
        'Kesulitan berat dalam mengunyah atau makan':1
    })
    
    data['Masalah_Kognitif']=data['Masalah_Kognitif'].map({
        'Tidak ada masalah daya ingat':3,
        'Ada sedikit masalah daya ingat':2,
        'Masalah daya ingat yang signifikan':1
    })

    # Seleksi hanya kolom numerik
    data = data.drop(columns=['Nama'])

    return data
# Fungsi untuk menghitung centroid
def calculate_centroids(scaled_data, clusters):
    centroids = []
    for cluster in clusters:
        if len(cluster) > 0:
            centroid = np.mean(scaled_data[cluster], axis=0)
            centroids.append(centroid)
    return np.array(centroids)
    
# Proses data baru
data_baru_numeric = preprocess_data(data_baru)  # Pastikan fungsi ini ada dan bekerja dengan baik
scaler = joblib.load('scaler.pkl')  # Muat scaler
data_baru_scaled = scaler.transform(data_baru_numeric)  # Gunakan transform untuk data baru

# Muat model CURE yang telah disimpan
cure_instance = joblib.load('cure_model.pkl')
# Memprediksi kluster untuk data baru
clusters = cure_instance.get_clusters()  # Mendapatkan kluster dari model CURE
centroids = calculate_centroids(scaled_data, clusters)  # Hitung centroid jika perlu

# Fungsi untuk memprediksi kluster untuk data baru
def predict_cure(centroids, data_baru):
    distances = cdist([data_baru], centroids, metric='euclidean')
    return np.argmin(distances)

# Prediksi kluster untuk data baru
cluster_predicted = predict_cure(centroids, data_baru_scaled[0])
print(f"Data baru masuk ke dalam klaster: {cluster_predicted}")

```
