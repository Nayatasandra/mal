```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import AgglomerativeClustering
import joblib
```

```python
# Baca data CSV
data = pd.read_csv('dataset.csv')

```

```python
# Tampilkan beberapa baris awal data
print("Head dari data awal:")
print(data.head())
```

```python

# Lihat deskripsi statistik dari data
print("\nDeskripsi Statistik:")
print(data.describe())
```

```python

# Visualisasikan distribusi berat badan
plt.figure(figsize=(10,6))
sns.histplot(data['Berat_Badan'], kde=True, color='blue')
plt.title('Distribusi Berat Badan')
plt.xlabel('Berat Badan')
plt.ylabel('Frekuensi')
plt.show()
```

```python
# Mengonversi kolom teks menjadi numerik (untuk kolom frekuensi makan, variasi makanan, dsb.)
data['Jenis_Kelamin'] = data['Jenis_Kelamin'].map({'Perempuan': 1, 'Laki-laki': 0})
data['Penyakit_Kronis'] = data['Penyakit_Kronis'].map({'Ya': 1, 'Tidak': 0})
data['Stres'] = data['Stres'].map({'Ya': 1, 'Tidak': 0})
data['Frekuensi_Makan'] = data['Frekuensi_Makan'].map({
    'Tiga kali atau lebih':3,
    'Dua kali': 2,
    'Satu kali atau kurang': 1
})

data['Variasi_Makanan'] = data['Variasi_Makanan'].map({
    'Makanan bervariasi (misalnya daging, sayuran, buah)': 3,
    'Makanan terbatas pada beberapa jenis saja': 2,
    'Makanan sangat terbatas atau monoton': 1
})

data['Asupan_Protein'] = data['Asupan_Protein'].map({
    'Dua kali sehari atau lebih': 3,
    'Sekali sehari': 2,
    'Kurang dari sekali sehari atau tidak pernah':1
})

data['Mobilitas']=data['Mobilitas'].map({
    'Bergerak dengan bebas tanpa bantuan':3,
    'Bergerak dengan bantuan':2,
    'Tidak dapat bergerak tanpa bantuan orang lain':1
})

data['Aktivitas_Sehari_hari']=data['Aktivitas_Sehari_hari'].map({
    'Tidak ada kesulitan':3,
    'Kesulitan ringan':2,
    'Kesulitan berat':1
})

data['Kesehatan_Mulut']=data['Kesehatan_Mulut'].map({
    'Tidak ada masalah':4,
    'Ada kesulitan ringan':3,
    'Kesulitan berat':2,
    'Kesulitan berat dalam mengunyah atau makan':1
})

data['Masalah_Kognitif']=data['Masalah_Kognitif'].map({
    'Tidak ada masalah daya ingat':3,
    'Ada sedikit masalah daya ingat':2,
    'Masalah daya ingat yang signifikan':1
})
```

```python
# Tampilkan data yang sudah dikonversi
print("\nData setelah konversi kategorikal:")
print(data.head())
```

```python
# Isi missing values (jika ada)
data.fillna(data.mean(), inplace=True)
```

```python
# Normalisasi data (kecuali kolom 'Nama')
scaler = StandardScaler()
scaled_data = scaler.fit_transform(data.drop(columns=['Nama']))
```

```python
# Tampilkan data yang sudah dinormalisasi
print("\nData setelah normalisasi (5 baris pertama):")
print(pd.DataFrame(scaled_data, columns=data.columns[1:]).head())
```

```python
# Membuat model clustering dengan 3 kluster
clustering_model = AgglomerativeClustering(n_clusters=3)
clustering_model.fit(scaled_data)
```

```python
# Tambahkan hasil clustering ke data
data['Cluster'] = clustering_model.labels_
```

```python
# Tampilkan hasil clustering
print("\nData dengan klaster:")
print(data[['Nama', 'Cluster']].head(10))
```

```python
# Tampilkan hasil clustering
print("\nData dengan klaster:")
print(data[['Nama', 'Cluster']].tail(10))
```

```python
# Visualisasikan klaster berdasarkan berat badan dan asupan protein
plt.figure(figsize=(10,6))
sns.scatterplot(x=data['Berat_Badan'], y=data['Asupan_Protein'], hue=data['Cluster'], palette='deep')
plt.title('Klaster berdasarkan Berat Badan dan Asupan Protein')
plt.xlabel('Berat Badan')
plt.ylabel('Asupan Protein')
plt.show()
```

```python
# Simpan model ke file
joblib.dump(clustering_model, 'clustering_model.pkl')
print("Model telah disimpan ke clustering_model.pkl")
```

```python
import joblib
import pandas as pd
from sklearn.preprocessing import StandardScaler

clustering_model = joblib.load('clustering_model.pkl')

# Muat data baru
data_baru = pd.DataFrame({
    'Nama': ['Hasan'],
    'Umur': [75],
    'Jenis_Kelamin': ['Laki-laki'],
    'LiLA': [30],
    'Berat_Badan': [50],
    'Tinggi_Badan': [160],
    'Frekuensi_Makan': ['Dua kali'],
    'Variasi_Makanan': ['Makanan bervariasi (misalnya daging, sayuran, buah)'],
    'Asupan_Protein': ['Dua kali sehari atau lebih'],
    'Mobilitas': ['Bergerak dengan bebas tanpa bantuan'],
    'Aktivitas_Sehari_hari': ['Tidak ada kesulitan'],
    'Kesehatan_Mulut': ['Tidak ada masalah'],
    'Penyakit_Kronis': ['Tidak'],
    'Stres': ['Tidak'],
    'Masalah_Kognitif': ['Ada sedikit masalah daya ingat']
})

# Preprocessing data baru (ubah kategorikal dan normalisasi)
def preprocess_data(data):
    data['Jenis_Kelamin'] = data['Jenis_Kelamin'].map({'Perempuan': 1, 'Laki-laki': 0})
    data['Penyakit_Kronis'] = data['Penyakit_Kronis'].map({'Ya': 1, 'Tidak': 0})
    data['Stres'] = data['Stres'].map({'Ya': 1, 'Tidak': 0})
    data['Frekuensi_Makan'] = data['Frekuensi_Makan'].map({
        'Tiga kali atau lebih':3,
        'Dua kali': 2,
        'Satu kali atau kurang': 1
    })
    
    data['Variasi_Makanan'] = data['Variasi_Makanan'].map({
        'Makanan bervariasi (misalnya daging, sayuran, buah)': 3,
        'Makanan terbatas pada beberapa jenis saja': 2,
        'Makanan sangat terbatas atau monoton': 1
    })
    
    data['Asupan_Protein'] = data['Asupan_Protein'].map({
        'Dua kali sehari atau lebih': 3,
        'Sekali sehari': 2,
        'Kurang dari sekali sehari atau tidak pernah':1
    })
    
    data['Mobilitas']=data['Mobilitas'].map({
        'Bergerak dengan bebas tanpa bantuan':3,
        'Bergerak dengan bantuan':2,
        'Tidak dapat bergerak tanpa bantuan orang lain':1
    })
    
    data['Aktivitas_Sehari_hari']=data['Aktivitas_Sehari_hari'].map({
        'Tidak ada kesulitan':3,
        'Kesulitan ringan':2,
        'Kesulitan berat':1
    })
    
    data['Kesehatan_Mulut']=data['Kesehatan_Mulut'].map({
        'Tidak ada masalah':4,
        'Ada kesulitan ringan':3,
        'Kesulitan berat':2,
        'Kesulitan berat dalam mengunyah atau makan':1
    })
    
    data['Masalah_Kognitif']=data['Masalah_Kognitif'].map({
        'Tidak ada masalah daya ingat':3,
        'Ada sedikit masalah daya ingat':2,
        'Masalah daya ingat yang signifikan':1
    })

    # Seleksi hanya kolom numerik
    data = data.drop(columns=['Nama'])

    return data

# Proses data baru
data_baru_numeric = preprocess_data(data_baru)

# Normalisasi data baru
scaler = StandardScaler()
data_baru_scaled = scaler.fit_transform(data_baru_numeric)

# Prediksi cluster untuk data baru
cluster_predicted = clustering_model.predict(data_baru_scaled)

# Menampilkan hasil cluster
print(f"Data baru masuk ke dalam klaster: {cluster_predicted[0]}")
```
